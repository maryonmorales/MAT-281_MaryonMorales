El **cuarteto de Anscombe** comprende cuatro conjuntos de datos que tienen las mismas propiedades estadísticas, pero que evidentemente son distintas al inspeccionar sus gráficos respectivos.

Cada conjunto consiste de once puntos (x, y) y fueron construidos por el estadístico F. J. Anscombe. El cuarteto es una demostración de la importancia de mirar gráficamente un conjunto de datos antes de analizarlos.

3- Ahora vemos el ajuste lineal y como se ve graficamente

```python
import statsmodels.api as sm
from sklearn import datasets
from sklearn.model_selection import train_test_split

# import some data to play with

X1 = grupo1[['x']] # we only take the first two features.
y1 = grupo1['y']

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

model_1 = sm.OLS(y_train1, sm.add_constant(X_train1))
results_1 = model_1.fit()

sns.lmplot(
    x='x',
    y='y',
    data=grupo1,
    height = 8,
)  
plt.show()
```

```python
X2 = grupo2[['x']] # we only take the first two features.
y2 = grupo2['y']

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42)

model_2 = sm.OLS(y_train2, sm.add_constant(X_train2))
results_2 = model_2.fit()
```

```python


sns.lmplot(
    x='x',
    y='y',
    data=grupo2,
    height = 8,
)  
plt.show()
```

```python
X3 = grupo3[['x']] # we only take the first two features.
y3 = grupo3['y']

X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=42)

model_3 = sm.OLS(y_train3, sm.add_constant(X_train3))
results_3 = model_3.fit()
```

```python
sns.lmplot(
    x='x',
    y='y',
    data=grupo3,
    height = 8,
)  
plt.show()
```

```python
X4 = grupo4[['x']] # we only take the first two features.
y4 = grupo4['y']

X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, y4, test_size=0.2, random_state=42)

model_4 = sm.OLS(y_train4, sm.add_constant(X_train4))

results_4 = model_4.fit()
```
