```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
total_profit=df["total_profit"]

meses = df["month_number"]
# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar
plt.plot(
    meses, # eje x
    total_profit, # eje y
    label='linea', # etiquetado
    color="red", # color
    linewidth=1 # tamano de la curva
)
plt.legend() # agregar etiquetado
plt.title("Ganancia Total a través de los meses") # agregar titulo
plt.xlabel('Meses') # nombre eje x
plt.ylabel('Ganancia total') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

```python
total_profit=df["total_profit"]

meses = df["month_number"]

fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    meses, # eje x
    total_profit, # eje y
    s=10, # tamano de los puntos
    c="blue", # color
    alpha=1, # intensidad color
    marker='o', # forma de los puntos
    label="punto" # etiquetdo fijando posicion
)
plt.xlabel("Meses")
plt.ylabel("Ganancia Total")
plt.legend(loc='upper left')
plt.title("Ganancia total a través de los meses")
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
meses = df["month_number"]
facecream= df["facecream"]
facewash=df["facewash"]
toothpaste=df["toothpaste"]
bathingsoap=df["bathingsoap"]
shampoo=df["shampoo"]
moisturizer=df["moisturizer"]


# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar

# a) facecream
plt.plot(
    meses, # eje x
    facecream, # eje y
    label='facecream', # etiquetado
    color="darkviolet", # color
    linewidth=2 # tamano de la curva
)
# b) facewash
plt.plot(
    meses, # eje x
    facewash, # eje y
    label='facewash', # etiquetado
    color="b", # color
    linewidth=2 # tamano de la curva
)
# c) toothpaste 
plt.plot(
    meses, # eje x
    toothpaste, # eje y
    label='toothpaste', # etiquetado
    color="r", # color
    linewidth=2 # tamano de la curva
)

# d) bathingsoup 
plt.plot(
    meses, # eje x
    bathingsoap, # eje y
    label='bathingsoap', # etiquetado
    color="green", # color
    linewidth=2 # tamano de la curva
)

# e) shampoo 
plt.plot(
    meses, # eje x
    shampoo, # eje y
    label='shampoo', # etiquetado
    color="gold", # color
    linewidth=2 # tamano de la curva
)

# f) moisturizer 
plt.plot(
    meses, # eje x
    moisturizer, # eje y
    label='moisturizer', # etiquetado
    color="c", # color
    linewidth=2 # tamano de la curva
)

plt.legend() # agregar etiquetado
plt.title("Productos vendidos a lo largo de los meses") # agregar titulo
plt.xlabel('Meses') # nombre eje x
plt.ylabel('Productos') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico


```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
meses = df["month_number"]
facecream= df["facecream"]
facewash=df["facewash"]


x = meses  # the label locations
width = 0.4  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, facecream, width, label='Facecream')
rects2 = ax.bar(x + width/2, facewash, width, label='Facewash')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Productos vendidos')
ax.set_title('Facecream y Facewash vendidos a lo largo de los meses')
ax.set_xlabel('Meses')
ax.set_xticks(x)
ax.legend()






```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
# Pre-format DataFrame
stats_df = df.drop(["total_units","total_profit","month_number"], axis=1)
 
# New boxplot using stats_df
plt.figure(figsize=(10, 6))
sns.boxplot(data=stats_df)
plt.title("Ventas de productos")
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
#sumas de cada uno producto

facecream= df["facecream"].sum()
facewash=df["facewash"].sum()
toothpaste=df["toothpaste"].sum()
bathingsoap=df["bathingsoap"].sum()
shampoo=df["shampoo"].sum()
moisturizer=df["moisturizer"].sum()




labels = 'Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap','Shampoo','Moisturizer'
sizes = [facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer]
explode = (0, 0, 0, 0.05, 0, 0)  # solo "explota" Bathingsoap (ya que es el más vendido)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', 
        shadow=False, startangle=90)
ax1.axis('equal') 
plt.title("Ventas realizadas el año pasado")
plt.show()
```
